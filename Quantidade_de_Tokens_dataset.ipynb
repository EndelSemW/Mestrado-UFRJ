```python
import openai
from getpass import getpass

api_key = getpass('Digite sua chave da API da OpenAI aqui: ')
openai.api_key = api_key

def count_exact_tokens(text):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "Your task is to analyze the text and count the tokens."},
            {"role": "user", "content": text}
        ],
        max_tokens=1
    )
    return response['usage']['total_tokens']

total_tokens = df['text'].apply(count_exact_tokens).sum()

print("Total exato de tokens na coluna 'text':", total_tokens)



```
